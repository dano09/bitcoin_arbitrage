```python
bitcoin_data = pd.read_csv('bitcoin_data_min_tick.csv')
bitcoin_data = bitcoin_data.set_index('timestamp', drop=False)
```

```python
news_data = pd.read_csv('news_data_min_tick.csv')
news_data = news_data.set_index('published_at', drop=False)
del(news_data['source.1'])
news_data.head()
```

```python
news_data_with_predictions = pd.read_csv('test_finished_product.csv')
news_data_with_predictions = news_data_with_predictions.set_index('published_at', drop=False)

# Format sentiment (float -> Int)
news_data_with_predictions.sentiment = news_data_with_predictions.sentiment.astype(int)
#df.col = df.col.astype(int)
news_data_with_predictions.head()
```

## Part 2 Combine dataframes 

```python
#news_data.head()
bitcoin_data['article_published'] = 0
bitcoin_data.head()
#temp = pd.concat([bitcoin_data, news_data], axis=1, join='left')

# Used for initial graph to see where 
# Caution - Removes articles for which there are no timestamps  (~200 of 7500)
bitcoin_data.loc[news_data_with_predictions.index, 'article_published'] = 1


bitcoin_data.head()
#bitcoin_data.shape #106561
#news_data.shape #7747
#news_data_with_predictions.shape #3786
```

## Part 3 Graph

```python
# Get all dates for widget
available_dates = list(set(news_data_with_predictions.published_at.str.slice(0, 10)))
available_dates.sort()
available_dates
```

```python
# Function to call when date is chosen
def choose_date(user_date):
    
    # Filter bitcoin time-series based on date
    plot_bitcoin_data = bitcoin_data[bitcoin_data['timestamp'].str.contains(user_date)]
    
    # Get timestamps for articles for user_date from news dataframe 
    plot_news_data = news_data_with_predictions[news_data_with_predictions['published_at'].str.contains(user_date)]
    
    # Plot Bitcoins Price
    trace0 = go.Scatter(
        x=plot_bitcoin_data.timestamp, 
        y=plot_bitcoin_data.price, 
        name='BTC/USD')
    
    temp_bitcoin = plot_bitcoin_data[plot_bitcoin_data['article_published'] == 1]
    
    # Plot Points of News Articles
    trace1 = go.Scatter(
        x=temp_bitcoin.timestamp, 
        y=temp_bitcoin.price, 
        mode='markers',
        name='Article',
        marker = dict( 
            size=5, 
           
            line = dict(width=1)
        ),
        #text=plot_news_data.loc[temp_bitcoin.timestamp]['description'],
        #hoverinfo='text',
        textposition='top')
        #hoverlabel = {'bgcolor'='blue'})
    
    #print(min(plot_bitcoin_data.price))
    #print(plot_bitcoin_data)
    text_line = min(plot_bitcoin_data.price)
    print('text_line is: {}'.format(text_line))

    # Plot news Info
    trace2 = go.Scatter(
        x=temp_bitcoin.timestamp,
        y=[min(plot_bitcoin_data.price) - 10]*len(temp_bitcoin.timestamp),
        #mode='markers',
        name='Article',
        text=plot_news_data.loc[temp_bitcoin.timestamp]['title'] + '<br>' + 
        plot_news_data.loc[temp_bitcoin.timestamp]['source'] ##+ Image(url,width=300, height=300)
        #textposition='bottom'
    )

    
    #print(trace1.hoverinfo)
    
    plot_data = [trace0, trace1, trace2]#, trace2]
    #layout= go.Layout(title = 'Styled Scatter',
    #                  images= [dict(
    #                  source= "https://images.plot.ly/language-icons/api-home/python-logo.png",
    #                  xref= "paper", yref= "paper", x=0, y=0, sizex=10, sizey= 10,
    #                  sizing= "contain", xanchor="center", yanchor="top", opacity= 0.5,
    #                  layer= "above")])

    
    fig = dict(data=plot_data, layout=layout)
    
    #py.iplot(fig)
    iplot([trace0])

# Create dropdown widget
date_dropdown = widgets.Dropdown(
    options=available_dates,
    description='Choose Date:',
    disabled=False,
)

interact(choose_date, user_date=date_dropdown)

#Image(url,width=300, height=300)
#print(article)
#shapes.append({
#    'type': 'marker',
#    'x0': article_time,
#    'x1': article_time,
#    'y0': min_y0,
#    'y1': max_y0,
#    'line': { 'width': 1}
#})
#print(article['published_at'])
```
